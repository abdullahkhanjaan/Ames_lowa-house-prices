<font color='amber'>4.1 MISSING VALUES ANALYSIS</font>

```python
#cols_with_missing = [col for col in train.columns if train[col].isnull().any()]
cols_with_missing = ames_data.isnull().sum()
cols_with_missing = cols_with_missing[cols_with_missing>0]
cols_with_missing.sort_values(inplace=True)
fig, ax = plt.subplots(figsize=(7,6))  
width = 0.70 # the width of the bars 
ind = np.arange(len(cols_with_missing))  # the x locations for the groups
ax.barh(ind, cols_with_missing, width, color="red")
ax.set_yticks(ind+width/2)
ax.set_yticklabels(cols_with_missing.index, minor=False)
plt.xlabel('Count')
plt.ylabel('Features')
```

```python
# Count and percentage of  missing values in testing data
print('Percentage of missing values in each columns')
#test.apply(lambda x: round((x.isna().sum()/len(x))*100,1))
total = ames_data.isnull().sum().sort_values(ascending=False)
percent = (ames_data.isnull().sum()/ames_data.isnull().count()).sort_values(ascending=False)
missing_data_te = pd.concat([total, round(percent*100,2)], axis=1, keys=['Total', 'Percent'])
missing_data_te[missing_data_te.Total>=1]
```

<font color='aqua'>Missing value analysis shows that most of the columns of dataset contain null values. categoricaland numerical columns both types contain null values. categorical null values will be imputed with 'None' and numerical with 0 while there are some other variables like 'electrical','sales type, 'exterior', theses will be imputed using their modes. as we have seen above 'garageyrbuilt and 'yearbuilt' have multicollinearity. so 'garageyrbuilt will be imputed using 'yearbuilt'. On missing value analysis bases, varaibles are categorized to three groups, categorical, numerical, other. these three groups will be imputed as discussed above.</font>

```python
from sklearn.impute import SimpleImputer
# Group 1:
group_1 = [
    'Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu', 'Garage Type',
    'Garage Finish', 'Garage Qual', 'Garage Cond', 'Bsmt Qual', 'Bsmt Cond',
    'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2', 'Mas Vnr Type']
x[group_1] = x[group_1].fillna("None")
# Group 2:
group_2 = [
    'Garage Area', 'Garage Cars', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF',
    'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Mas Vnr Area']
x[group_2] = x[group_2].fillna(0)
# Group 3:
group_3a = [
    'Functional', 'MS Zoning', 'Electrical', 'Kitchen Qual', 'Exterior 1st',
    'Exterior 2nd', 'Sale Type', 'Utilities']
imputer = SimpleImputer(strategy='most_frequent')
x[group_3a] = pd.DataFrame(imputer.fit_transform(x[group_3a]), index=x.index)
x['Lot Frontage'] = x['Lot Frontage'].fillna(x['Lot Frontage'].mean())
x['Garage Yr Blt'] = x['Garage Yr Blt'].fillna(x['Year Built'])

sum(x.isnull().sum())
```

<font color='amber'>4.4.2 NORMALZING OTHER NUMERICAL VARIABLES USING LOG TRANSFORMATION</font>


```python
skewness = x.skew().sort_values(ascending=False)
skewness[abs(skewness) > 0.75]
```

```python
# List of skewed columns
skewed_cols = list(skewness[abs(skewness) > 0.5].index)
# Remove 'MSSubClass' and 'SalePrice'
skewed_cols = [col for col in skewed_cols if col not in ['MSSubClass', 'SalePrice']]
# Log-transform skewed columns
for col in skewed_cols:x[col] = np.log(1 + x[col])
```

<font color='aqua'>Variables having skewness less than or greater than 1 is called to be highly skewed, skewness btw -1 to -0.5 and 0.5 to 1 is moderate while skewness btw -0.5 and 0.5 are nearly symmetric. Skewness of the variables are calculated and then removed using log transformation.</font>


<font color='amber'>4.5 CREATE NEW VARAIBLES FROM WEAK VARIABLES</font>

```python
x['total SqFeet'] = x['Total Bsmt SF'] + x['1st Flr SF'] +x['2nd Flr SF']
x['totalBathroom'] = x['Full Bath'] + x['Bsmt Full Bath'] + 0.5 * (x['Half Bath'] + x['Bsmt Half Bath'])
x['houseAge'] = x['Yr Sold'] - x['Year Built']
x['reModeled'] = np.where(x['Year Remod/Add'] == x['Year Built'], 0, 1)
x['isNew'] = np.where(x['Yr Sold'] == x['Year Built'], 1, 0)
```
